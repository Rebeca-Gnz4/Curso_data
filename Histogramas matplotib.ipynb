## Plot descriptive statistics matplotib

```python
import pandas as pd

# Import the air quality dataset 
CO2Data = pd.read_csv("data/Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(CO2Data.info())
```

- Plot histogram of DIC

```python
import matplotlib.pyplot as plt

# Select data DIC
dic_data = CO2Data["dic_micromol_kg"]

plt.hist(dic_data, bins=10, color="green")
plt.xlabel("DIC (µmol/kg)")
plt.ylabel("Frequency")
plt.show()
```

- Plot histogram of TA

```python
ta_data = CO2Data["ta_micromol_kg"]

# Select data TA
plt.hist(ta_data, bins=10, color='lightblue')
plt.xlabel("TA (µmol/kg)")
plt.ylabel("Frequency")
plt.show()
```
